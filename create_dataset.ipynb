```python
import pandas as pd
import os
import shutil
```

```python
chx_filepath = "../covid-chest-xray-dataset/metadata.csv"
chx_imagepath = "../covid-chest-xray-dataset/images"
```

```python
chx_df = pd.read_csv(chx_filepath)
```

```python
chx_df.head()
```

```python
target_dir = "../dataset/COVID-POSITIVE"

if not os.path.exists(target_dir):
    os.mkdir(target_dir)
    
```

```python
counter = 0
for (i, row) in chx_df.iterrows():
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"] == "PA":
        filename = row["filename"]
        imagepath_chx = os.path.join(chx_imagepath, filename)
        copy_imagepath = os.path.join(target_dir, filename)
        shutil.copy2(imagepath_chx, copy_imagepath)
        counter += 1
        
print(counter)
```

```python
normal_image_path = "../COVID-19-Radiography-Database/NORMAL"
```

```python
target_dir = "../dataset/NORMAL"

if not os.path.exists(target_dir):
    os.mkdir(target_dir)
```

```python
import random

image_names = os.listdir(normal_image_path)
random.shuffle(image_names)

for i in range(196):
    image_name = image_names[i]
    print(image_name)
    image_path = os.path.join(normal_image_path, image_name)
    target_path = os.path.join(target_dir, image_name)
    shutil.copy2(image_path, target_path)
```
